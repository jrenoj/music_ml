```python
def preprocess_data(df):
    df_cleaned = df.dropna()
    return df_cleaned

data=preprocess_data(data)

genre_data=preprocess_data(genre_data)

year_data=preprocess_data(year_data)
```

```python
#change features
features = ['acousticness', 'danceability', 'energy', 'instrumentalness', 'liveness', 'loudness', 'speechiness', 'tempo', 'valence','duration_ms','explicit','key','mode','year']
X = data[features]
y = data['popularity']
```

```python
df = pd.concat([X, y], axis=1)
correlation_matrix = df.corr()

plt.figure(figsize=(20, 20))
sns.heatmap(correlation_matrix, annot=True, cmap = 'coolwarm',  linewidths=0.5)

plt.title('Feature Correlation')
plt.show()

```

```python
# Compute correlations between each feature and the target 'popularity'
correlations = X.corrwith(y)

# Convert correlations to a DataFrame for plotting
corr_df = pd.DataFrame({'Feature': features, 'Correlation': correlations})

# Sort by the absolute value of the correlation for better visualization
corr_df = corr_df.reindex(corr_df['Correlation'].sort_values(ascending=False).index)


plt.figure(figsize=(12, 8))
sns.barplot(x='Correlation', y='Feature', color= "green", data=corr_df, palette='coolwarm')

plt.title('Feature Correlation with Popularity')
plt.show()
```

```python
data['half a decade'] = (data['year'] // 5 * 5).astype(str) + "~" + (data['year'] // 5 * 5 + 4).astype(str)


sns.set_theme(rc={'figure.figsize': (11, 6)})
sns.countplot(data=data, x='half a decade', order=sorted(data['half a decade'].unique()))
plt.title('Count of Songs per each five years')
plt.xticks(rotation=45)
plt.show()
```

```python
#Consider changing here
music_features = ['acousticness', 'danceability', 'energy', 'instrumentalness', 'liveness',   'valence', 'speechiness']


def plot_features(features, title):
    plt.figure(figsize=(12, 8))
    for feature in features:
        plt.plot(year_data['year'], year_data[feature], label=feature)
    plt.title(title, fontsize=16)
    plt.xlabel('Year', fontsize=14)
    plt.ylabel('Feature Value', fontsize=14)
    plt.legend(title='Music Features', fontsize=12)
    plt.grid(True)
    plt.show()

plot_features(music_features, 'Trends of Music Features Over Time')
```

```python

genres_20 = genre_data.nlargest(20, 'popularity')
#Change features
#features = ['valence', 'energy', 'danceability', 'acousticness']
features = ['acousticness', 'danceability', 'energy', 'instrumentalness', 'liveness',   'valence', 'speechiness']
x = np.arange(len(genres_20['genres'])) 
width = 0.2 


plt.figure(figsize=(12, 6))

for i, feature in enumerate(features):
    plt.bar(x + i * width, genres_20[feature], width=width, label=feature)

plt.xticks(x + width * (len(features) - 1) / 2, genres_20['genres'], rotation=45, ha='right', fontsize=10)
plt.title('Top 20 Genres by Popularity and their Sound Features', fontsize=16)
plt.xlabel('Genres', fontsize=14)
plt.ylabel('Feature Values', fontsize=14)
plt.legend(title='Features', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```
